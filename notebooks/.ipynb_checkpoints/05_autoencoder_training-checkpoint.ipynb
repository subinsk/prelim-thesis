# Self-Supervised Learning using Convolutional Autoencoder

This notebook trains a convolutional autoencoder on benign
container image patches to learn a representation of normal
container filesystem structure.


```python
import os
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

PROJECT_ROOT = os.path.abspath("..")

PATCH_DIR = os.path.join(PROJECT_ROOT, "data", "processed", "patches")
```

```python
alpine_patch_path = os.path.join(
    PATCH_DIR, "alpine_3_19_patches.npy"
)

# Load patches again (fresh)
patches = np.load(alpine_patch_path)

print("Before:", patches.dtype, patches.min(), patches.max())

# Normalize AND cast explicitly
patches = patches.astype(np.float32) / 255.0
patches = np.expand_dims(patches, axis=1)  # (N, 1, 256, 256)

print("After:", patches.dtype, patches.min(), patches.max())
```

```python
# Load patches fresh
patches = np.load(alpine_patch_path)

print("Original shape:", patches.shape)
print("Original dtype:", patches.dtype)

# Ensure patches are (N, 256, 256)
if patches.ndim == 4:
    # case: (N, 1, 256, 256) → squeeze channel
    patches = np.squeeze(patches, axis=1)

elif patches.ndim == 5:
    # case: (N, 1, 1, 256, 256) → squeeze both
    patches = np.squeeze(patches, axis=(1, 2))

print("After squeeze:", patches.shape)

# Normalize and add SINGLE channel dimension
patches = patches.astype(np.float32) / 255.0
patches = np.expand_dims(patches, axis=1)

print("Final shape:", patches.shape)
```

```python
tensor_data = torch.from_numpy(patches)

dataset = TensorDataset(tensor_data)

loader = DataLoader(
    dataset,
    batch_size=16,
    shuffle=True
)

# HARD sanity check
batch = next(iter(loader))[0]
print("Batch shape going into model:", batch.shape)
```

```python
class ConvAutoencoder(nn.Module):
    def __init__(self):
        super().__init__()

        self.encoder = nn.Sequential(
            nn.Conv2d(1, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU()
        )

        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 1, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

model = ConvAutoencoder().to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
epochs = 5
losses = []

for epoch in range(epochs):
    epoch_loss = 0.0

    for (batch,) in loader:
        batch = batch.to(device)

        recon = model(batch)
        loss = criterion(recon, batch)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    avg_loss = epoch_loss / len(loader)
    losses.append(avg_loss)

    print(f"Epoch {epoch+1}/{epochs} - Loss: {avg_loss:.6f}")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(losses, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Reconstruction Loss")
plt.title("Autoencoder Training Loss (Benign Alpine Patches)")
plt.grid(True)
plt.show()
```

```python
model.eval()

with torch.no_grad():
    sample = tensor_data[0:1].to(device)
    recon = model(sample)

orig_img = sample.cpu().numpy()[0, 0]
recon_img = recon.cpu().numpy()[0, 0]

plt.figure(figsize=(6, 3))

plt.subplot(1, 2, 1)
plt.imshow(orig_img, cmap="gray")
plt.title("Original Patch")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(recon_img, cmap="gray")
plt.title("Reconstructed Patch")
plt.axis("off")

plt.show()
```

## Observations

- The autoencoder successfully learns to reconstruct benign container patches.
- Training loss decreases steadily, indicating learning of normal structure.
- Reconstructed patches preserve major structural patterns with minor smoothing.
- This confirms suitability of reconstruction error for anomaly detection.
