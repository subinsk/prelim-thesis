# Preliminary Thesis – Problem and Approach

## Problem Statement
Supervised malware detection models for Docker containers rely on labeled malware families.
In real-world scenarios, malware labels are often unavailable, incomplete, or outdated,
making such approaches ineffective against zero-day attacks.

## Motivation
Existing image-based container malware detection methods assume access to known malware
samples during training. This assumption does not hold in practical deployments where
novel or obfuscated malware appears continuously.

## Proposed Approach
This work explores a self-supervised anomaly detection framework for Docker containers.
The model is trained exclusively on benign container images to learn the normal structure
of container filesystems. Containers that deviate from this learned representation are
flagged as suspicious.

## High-Level Pipeline
Docker Image → Filesystem Tarball → Byte-to-Image Representation → 
Self-Supervised Model → Anomaly Score
